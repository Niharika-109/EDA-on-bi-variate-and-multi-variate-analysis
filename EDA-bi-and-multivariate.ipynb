```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
iris=sns.load_dataset('iris')
```

```python
flight=sns.load_dataset('flights')
```

```python
titanic=pd.read_csv('train.csv')
```

```python
# bivariate analysis 
#1.Scatterplot(numerical-numerical)
sns.scatterplot(df['total_bill'],df['tip'])
```

```python
#multivariate analysis
sns.scatterplot(df['total_bill'],df['tip'],hue=df['sex'])
sns.scatterplot(df['total_bill'],df['tip'],hue=df['sex'],size=df['size'])
sns.scatterplot(df['total_bill'],df['tip'],hue=df['sex'],style=df['smoker'])
```

```python
#bar plot(numerical-categorical)
#bivariate analysis
sns.barplot(titanic['Pclass'],titanic['Age'])
# black color line is confidence interval
```

```python
#multivate analysis
sns.barplot(titanic['Pclass'],titanic['Age'],hue=titanic['Sex'])
```

```python
#3.box plot(numerical-categorical)
sns.boxplot(titanic['Sex'],titanic['Age'])
```

```python
#multivariate
sns.boxplot(titanic['Sex'],titanic['Age'],hue=titanic['Survived'])
```

```python
#4.distplot(numerical-categorical)
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
#5.heatmap(categorical-categorical)
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
#to know the percentage we 
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

```python
#6.clustermap(categorical-categorical)
#it says closeness between the values
#tendo grams
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

```python
#7.pairplot
#it finds the numerical data and draw a plot
#grid of scatterplots
sns.pairplot(iris)
```

```python
#multivariate analysis
sns.pairplot(iris,hue='species')
```

```python
#line plot(numerical-numerical)
#if x axis has time based data then we use line plot
#similar to scatter plot
new=flight.groupby('year').sum().reset_index()
```

```python
sns.lineplot(new['year'],new['passengers'])
```

```python
sns.heatmap(flight.pivot_table(values='passengers',index='month',columns='year'))
```
